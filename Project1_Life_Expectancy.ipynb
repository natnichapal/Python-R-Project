## <a id="sub8"></a>8. Simple Linear Regression

The Adjusted R-squared value of 0.4787 indicates that the model explains 48% of the variance in the data; approximately 6% of variation in life expectancy can be explained by adult mortality.  p-value < 0.05 (significant) We have strong evidence that there is, in fact, a relationship between life expectancy and adult mortality. Eventhough R-squared value is not very high, we have statistically significant independent variable. Adult mortality is correlated with life expectancy, but it doesn't explain much of the variability in the life expectancy. Adult mortality is negatively related to life expectancy due to its negative slope. Therefore, we can conclude that adult mortality can be used to predict life expectancy.

BMI is correlated with life expectancy, but it doesn't explain much of the variability in the life expectancy. BMI is positively related to life expectancy due to its positive slope. ## <a id="sub9"></a>9. Multiple Linear Regression

<b> 9.1 Relationship between adult mortality, BMI, and life expectancy </b>

```r
m1 <- lm(Life.expectancy ~ Adult.Mortality + BMI, data = life)
summary(m1)
```

In this case, we can see the R-squared of 0.5813, approximately 58% of variation in life expectancy can be explained by adult mortality and BMI. The p-value for over all model is 2.2e-16. This is substantially lower than 0.05, we generally reject the null hypothesis. In other words, with a p-value less than 0.05, we have strong evidence that there is, in fact, a relationship between the adult mortality, BMI and life expectancy. 

<b> 9.2 Relationship between adult mortality, BMI, GDP, and life expectancy </b>

```r
m2 <- lm(Life.expectancy ~ Adult.Mortality + BMI + GDP, data = life)
summary(m2)
```

In this case, we can see the R-squared of 0.6334, approximately 63% of variation in life expectancy can be explained by adult mortality, BMI and GDP. The p-value for over all model is 2.2e-16. This is substantially lower than 0.05, we generally reject the null hypothesis. In other words, with a p-value less than 0.05, we have strong evidence that there is, in fact, a relationship between the adult mortality, BMI, GDP and life expectancy. 

Next, I use `anova( )` to compare m1 and m2. This will tell us whether adding GDP on top of adult mortality and BMI further improves the model.

```r
anova(m1,m2)
```

The result shows a significant result (p = 8.512502e-43). Thus, adding GDP further improves our model.

<b> 9.3 Relationship between adult mortality, BMI, GDP, percentage expenditure and life expectancy </b>

```r
m3 <- lm(Life.expectancy ~ Adult.Mortality + BMI + GDP + percentage.expenditure, data = life)
summary(m3)
```

In this case, we can see the R-squared of 0.6337, approximately 63% of variation in life expectancy can be explained by adult mortality, BMI, GDP, and percentage expenditure. The p-value for over all model is 2.2e-16. This is substantially lower than 0.05, we generally reject the null hypothesis. In other words, with a p-value less than 0.05, we have strong evidence that there is, in fact, a relationship between four factors and life expectancy. Perhaps the percentage expenditure variable is unnecessary in the model given its high p-value of 0.0124 which indicates that the percentage expenditure doesn't help explain the model very well.

Next, I use `anova( )` to compare m2 and m3. This will tell us whether adding percentage expenditure on top of adult mortality, BMI and GDP further improves the model.

```r
anova(m2,m3)
```

The result shows a non-significant result (p = 	0.123739). Thus, adding percentage expenditure doesn't further improves our model.

<b> 9.4 Relationship between adult mortality, BMI, GDP, infant deaths and life expectancy </b>

```r
m4 <- lm(Life.expectancy ~ Adult.Mortality + BMI + GDP + infant.deaths, data = life)
summary(m4)
```

In this case, we can see the R-squared of 0.6372, approximately 63.7% of variation in life expectancy can be explained by adult mortality, BMI, GDP, and infant deaths. The p-value for over all model is 2.2e-16. This is substantially lower than 0.05, we generally reject the null hypothesis. In other words, with a p-value less than 0.05, we have strong evidence that there is, in fact, a relationship between the adult mortality, infant deaths, BMI, GDP, and life expectancy. 

Next, I use `anova( )` to compare m2 and m4. This will tell us whether adding infant deaths on top of adult mortality, BMI, GDP further improves the model.

```r
anova(m2,m4)
```

The result shows a significant result (p = 	6.38984e-05). Thus, adding infant deaths further improves our model.

<b> 9.5 Relationship between adult mortality, BMI, GDP, infant deaths, Hepatitis B and life expectancy </b>

```r
m5 <- lm(Life.expectancy ~ Adult.Mortality + BMI + GDP + infant.deaths 
         + Hepatitis.B, data = life)
summary(m5)
```

The p-value for over all model is 2.2e-16. This is substantially lower than 0.05, we generally reject the null hypothesis. Thus, the best model so far is 

`Life.expectancy = (6.814e+01) + (-3.670e-02) * Adult.Mortality + (1.230e-01) * BMI + (1.722e-04) * GDP + (-3.797e-03) * infant.deaths + (2.294e-02) * Hepatitis.B`

The <b>residual plot</b> is a graph that shows the residuals on the vertical axis and the independent variable on the horizontal axis. If the points in a residual plot are randomly dispersed around the horizontal axis, a linear regression model is appropriate for the data; otherwise, a non-linear model is more appropriate.

The residual plot shows a fairly random pattern. The plot patterns above are non-random (U-shaped and inverted U), suggesting a better fit for a non-linear model.

The <b>Normal Q-Q plot</b>, or quantile-quantile plot. A Q-Q plot is a scatterplot created by plotting two sets of quantiles against one another. Notice the points fall along a line in the middle of the graph, but curve off in the extremities. Normal Q-Q plots that exhibit this behavior means this data have more extreme values than would be expected if they truly came from a Normal distribution. 

The <b>Scale-Location plot </b>It’s also called <b>Spread-Location plot</b>. This plot shows if residuals are spread equally along the ranges of predictors. This is how you can check the assumption of equal variance (homoscedasticity). It’s good because we see a horizontal line with the residuals appear randomly spread.
